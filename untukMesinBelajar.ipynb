```python
import pandas as pd
# Baca file Excel
df = pd.read_excel('databenar.xlsx')  # Ganti 'nama_file.xlsx' dengan nama file Excel Anda

# Simpan ke file CSV
df.to_csv('databenar.csv', index=False)  # Menyimpan DataFrame ke dalam file CSV baru
```

```python
#memisakhan atribut dengan label dimana x itu atribut dan y itu label
X = df[['Jumlah','Konsumsi']]
y = df['Kelas']


#membagi dataset menjadi data latih dan data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.1, random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier

#menentukan  model decision tree
tree_model= DecisionTreeClassifier()

#melatih model menggunakan data latih dengan fungsi fit
tree_model= tree_model.fit(X_train,y_train)
```

```python
#evaluasi model untuk meguji seberapa baik model yang digunakan dan akurasi

from sklearn.metrics import accuracy_score

y_pred =  tree_model.predict(X_test)

accuracy_score =round(accuracy_score(y_pred, y_test), 3)

print('Accuracy: ',accuracy_score)
```

```python
print('Penggunaan :',tree_model.predict([[1202323,1232232323232322]])[0])
```

```python
from sklearn.tree import export_graphviz

# Menggunakan 'tree_model' sebagai model yang ingin divisualisasikan
export_graphviz(tree_model, out_file="Tree_Kelas.dot", class_names=["A", "B", "C", "D"],
                feature_names=X.columns, impurity=False, filled=True)
```
